```python
from pyswip import Prolog
```

```python
def get_lista_turnos():
    p = Prolog()
    p.consult('prueba_ia_basico.pl')
    Q = p.query('turno(F, J, L, N, LL)')
    l = [{'Fi':q['F'], 'Jug':q['J'], 'Lado':q['L'], 'Turno':q['N'], 'Libre':q['LL']} for q in Q if q['F'] is not 0]
    l.sort(key=lambda x: x['Turno'])
    return l
def separa(turnos):
    c1 = []
    c2 = []
    for i in turnos:
        if(i['Lado'] is 1):
            c1.append(i)
        elif(i['Lado'] is 2):
            c2.append(i)
        else:
            c1.append(i)
            c2.append(i)
    return c1,c2
def arma_cola_1(c1):
    cola = []
    primer = c1[0]
    c1 = c1[:1]
    if(c1['Libre'][])
    for i in c1:
        if(i['Libre'] is 1): cola.append(tuple(i['Fi'][::-1]))
        else: cola.append(tuple(i['Fi']))
    return cola
def arma_cola_2(c2):
    cola = []
    for i in c2[::-1]:
        if(i['Libre'] is 2): cola.append(tuple(i['Fi'][::-1]))
        else: cola.append(tuple(i['Fi']))
    return cola
def arma_colas():
    lista = get_lista_turnos()
    c1,c2 = separa(lista)
    cc1 = arma_cola_1(c1)
    cc2 = arma_cola_2(c2)
    return cc2[:-1]+cc1
def cola_valida(cola):
    
    for i in range(1,len(cola)):
        ficha1 = cola[i-1]
        ficha2 = cola[i]
        if(ficha1[1] is not ficha2[0]):
            return False
    return True
```
