```python
def crea_lista_turnos():
    with open('prueba_turnos_p.txt','r') as f:
        ls = [i[:-2] for i in f.readlines()]
    #ls = [i for i in ls]
    ls=[{'ficha': i[8:14], 
        'jugador': i[16:17],
        'lado':i[19:20],
        'numTurno':i[22:23],
        'ladoLibre':i[25:26]} for i in ls[:-1]]
    return ls
crea_lista_turnos()    
```
